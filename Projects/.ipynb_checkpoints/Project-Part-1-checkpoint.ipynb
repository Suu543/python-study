## Project - 1
#### 학생 10명의 정보를 포함하는 튜플 (학생이름, 학과명, 학번, (연, 월, 일), 평균성적)들을 리스트에 포함시킨 후, 파이썬 내장함수 sum(), max(), min(), len()을 사용하여 학생들의 성적 분포 (최대, 최소, 평균)을 파악하여 출력하는 함수 statistics_student_GAP() 함수를 구현하라.

```
students = 
("Kim", 'EE', 12345, (2000, 12, 25), 4.0),
("Lee", 'ME', 11345, (2019, 9, 25), 4.2),
("Park", 'CE', 10345, (2000, 12, 25), 4.3)
statistics_student_GAP ::
    - L_GPAs = [4.0, 4.2, 4.3]
    - num_students = 3
    - Statistics of GPAs : Max (4.3), Min (4.0), Avg (4,155555555555556)
```

````python
students = [
    ("Kim", 'EE', 12345, (2000, 12, 25), 4.0),
    ("Lee", 'ME', 11345, (2019, 9, 25), 4.2),
    ("Park", 'CE', 10345, (2000, 12, 25), 4.3),
]

def getGPAStats(data):
    scores = []
    gpa = 4
    for s in data:
        scores.append(s[gpa])
    

    return {
        "Scores": scores,
        "Max": max(scores),
        "Min": min(scores),
        "Avg": sum(scores) / len(data),
        "Num_of_Students": len(scores)
    }


def start(students, GPA):
    print("students = ")

    for student in students:
        print(student)

    print(f'''statistics_student_GPA ::
        - L_GPAs = {GPA["Scores"]},
        - num_students = {GPA["Num_of_Students"]},
        - Statistics of GPAs : Max {GPA["Max"]}, Min {GPA["Min"]}, Avg {GPA["Avg"]}
    ''' )
    
start(students, getGPAStats(students))
````

## Project - 2
정수형 난수 리스트의 정렬 및 경과시간 측정
1. 중복되지 않은 정수형 (int) 난수를 지정된 개수 (예: 10,000 ~ 1,000,000) 만큼 생성하는 함수 genBigRandList(L, n)을 구현하라. 이 함수에는 리스트 L과 리스트에 포함에 포함될 중복되지 않는 난수의 개수 n이 전달된다.
2. 주어진

````python
import random as rd
import time


def genBigRandList(L, n):
    L = list(range(1, n + 1))
    rd.shuffle(L)
    return L

def mergeSort(L):
    if (len(L) < 2):
        return L
        
    result = []
    mid = int(len(L) / 2)
    left = mergeSort(L[:mid])
    right = mergeSort(L[mid:])

    left_index = 0
    right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] > right[right_index]:
            result.append(right[right_index])
            right_index += 1
        else:
            result.append(left[left_index])
            left_index += 1
    
    result += left[left_index:]
    result += right[right_index:]

    return result

def printFormatFunc(L):
    for (index, value) in enumerate(L[:20]):
        print(value, end = " ")
        if (index == 9):
            print("\n")

    print("\n ......")
    for (index, value) in enumerate(L[len(L) - 20 :]):
        print(value, end=" ")
        if (index == 9):
            print("\n")
    print("\n")

def printListSample(L, state):
    if (state == "before"):
        
        print("Before mergeSort of L :")
        printFormatFunc(L)

    elif (state == "after"):
        print("After mergeSort of L :")
        printFormatFunc(L)

def start():
    rdlist = []
    list_length = input("Input size of random list L (-1) to quit) = ")
    random_list = genBigRandList(rdlist, int(list_length))
    printListSample(random_list, "before")
    measure_merge_start = time.time()
    mergeSort_list = mergeSort(random_list)
    measure_merge_end = time.time()
    printListSample(mergeSort_list, "after")
    print(f"mergeSort() for list L (size = {list_length}) took {measure_merge_end - measure_merge_start} sec")

start()
````
