### def keyword

- 함수를 정의할때 def keyword를 사용한다.

```python
def addTwoNumbers(인자1, 인자2):
    return 인자1 + 인자2
```

```python
first_result = addTwoNumbers(1, 2)
second_result = addTwoNumbers(5, 4)
```

```python
first_result
```

```python
second_result
```

```python
def addTwoNumbersAndSayHello(firstNum, secondNum, name):
    print("firstNum", firstNum)
    print("secondNum", secondNum)
    print("firstNum + secondNum = ", firstNum + secondNum)
    print("안녕...", name)
    return
```

```python
addTwoNumbersAndSayHello(1, 5, "상호")
```

```python
addTwoNumbersAndSayHello(100, 100, "용수")
```

```python
firstNum = 2
secondNum = 3
name = "Hello"
print("firstNum", firstNum)
print("secondNum", secondNum)
print("firstNum + secondNum = ", firstNum + secondNum)
print("안녕...", name)
```

```python
# a / b 한 값을 그냥 출력해도되지만, 이 값을 저장하고 싶은 경우
# 리턴값을 변수에 할당하고 싶은경우 리턴에 전달해준 값이 변수에 할당된다고 생각할 수 있다.
def divide(a, b):
    return a / b
```

```python
division_result = divide(10, 5)
```

```python
division_result
```

```python
def multiplicationTable(num):
    for i in range(1, 10):
        print(f"{num} * {i} = {num * i}")
    return 
```

```python
multiplicationTable(4)
```

```python
while True:
    stage = int(input("계산하고 싶은 단을 입력해주세요 - 멈추고 싶으면 0 입력: "))
        
    if stage == 0:
        print("계산을 종료하겠습니다...")
        break
        
    multiplicationTable(stage)
    
```
